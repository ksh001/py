##인천의 분양가 높은 순

```
df.loc[df['지역명']=='서울', ['지역명', '연도', '분양가']].sort_index(axis=0, ascending=False)[:20]
```

```
df2 = df.loc[df['지역명']=='인천',['지역명', '연도', '분양가']]
df2.sort_index(axis=0, ascending=False)[:20]
```

```
df.loc[(df['지역명']=='인천') & (df['연도']==2018),['지역명', '연도', '분양가']]
```
